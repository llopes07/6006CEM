```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
max_temp = [30,31,32,28,27,29,26]
min_temp = [23,22,20,24,18,19,10]
avg_temp = [25,28,28,26,23,25,24]
wind = [50,100,40,65,80,75,50]
```

```python
plt.title('Wind', fontsize=18, color='red')
plt.ylabel('Wind speed', color='green')
plt.xlabel('Days of the week', color='green')
plt.scatter(days, wind, color='red', s=[e*3 for e in wind], alpha=0.7)
plt.grid()
```

```python
plt.plot(days, avg_temp, label='avg')
plt.plot(days, min_temp, label='min')
plt.plot(days, max_temp, label='max')
plt.title('Temperature of the week', fontsize=18, color='red')
plt.ylabel('Temperature', color='green')
plt.xlabel('Days of the week', color='green')
plt.grid()
plt.legend()
```
