```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
cols = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']
df = pd.read_csv('datasets/iris.data', names=cols)
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['class'] = le.fit_transform(df['class'])
df
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
pca = PCA(n_components=2)
pComponent = pca.fit_transform(X)
pDF = pd.DataFrame(data=pComponent, columns=['PC1', 'PC2'])
df = pd.concat([pDF, df[['class']]], axis=1)
df
```

```python
pca.explained_variance_ratio_
```

```python
pca = PCA(n_components=3)
pComponent = pca.fit_transform(X)
pDF = pd.DataFrame(data=pComponent, columns=['PC1', 'PC2', 'PC3'])
df2 = pd.concat([pDF, df[['class']]], axis=1)
df2
```
