```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('datasets/mnist_train.csv')
df_train.head(3)
```

```python
df_test = pd.read_csv('datasets/mnist_test.csv')
df_test.head(3)
```

```python
X_train = df_train.iloc[:, 1:].values
X_test = df_test.iloc[:, 1:].values
y_train = df_train.iloc[:, 0].values
y_test = df_test.iloc[:, 0].values
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.fit_transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
def fitAndTest(X_train, X_test, y_train, y_test):
    model = MLPClassifier(random_state=0)
    model.fit(X_train, y_train)
    accuracy = model.score(X_train, y_train)
    print('Training score: {:.3f}'.format(accuracy))
    accuracy = model.score(X_test, y_test)
    print('Testing score: {:.3f}'.format(accuracy))
```

```python
print('Before scaling: ')
fitAndTest(X_train, X_test, y_train, y_test)
```

```python
print('After scaling: ')
fitAndTest(X_train_scaled, X_test_scaled, y_train, y_test)
```

```python
from sklearn.decomposition import PCA
pac = PCA(n_components=10)

pComponent = pac.fit_transform(X_train_scaled)
pDF = pd.DataFrame(data=pComponent, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10'])
df_train_2 = pd.concat([pDF,df_train['label']], axis=1)
df_train_2
```

```python
pComponent = pac.fit_transform(X_test_scaled)
pDF = pd.DataFrame(data=pComponent, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10'])
df_test_2 = pd.concat([pDF, df_test['label']], axis=1)
df_test_2
```

```python
X_train_2 = df_train_2.iloc[:, :-1].values
X_test_2 = df_test_2.iloc[:, :-1].values
y_train_2 = df_train_2.iloc[:, -1].values
y_test_2 = df_test_2.iloc[:, -1].values
```

```python
X_train_2_scaled = sc.fit_transform(df_train_2)
X_test_2_scaled = sc.fit_transform(df_test_2)

fitAndTest(X_train_2_scaled, X_test_2_scaled, y_train_2, y_test_2)
```
