```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numberOfStudents = [1400,600,300,410,250]
courseNames = ['CS','Games','COMP','EH','ITB']
```

```python
explode = (0,0.15,0,0,0)
plt.pie(numberOfStudents, labels=courseNames, explode=explode, autopct='%1.2f%%', shadow=True, startangle=45, radius=1.5)
```
